```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('covid_19_india.csv')
groupedby_acc = df.groupby('State/UnionTerritory')[['Deaths']].sum()
fig, ax = plt.subplots()
ax.plot(groupedby_acc.index, groupedby_acc['Deaths'])
labels = ax.get_xticklabels()
plt.title('Corona in 2019')
plt.xlabel('State/UnionTerritory')
plt.ylabel('Deaths')
plt.show()
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.stats import multivariate_normal
X = np.linspace(-5,5,50)
Y = np.linspace(-5,5,50)
X, Y = np.meshgrid(X,Y)
X_mean = 0; Y_mean = 0
X_var = 5; Y_var = 8
pos = np.empty(X.shape+(2,))
pos[:,:,0]=X
pos[:,:,1]=Y
rv = multivariate_normal([X_mean, Y_mean],[[X_var, 0], [0, Y_var]])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos), cmap="plasma")
plt.show()
```
